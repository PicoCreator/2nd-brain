```python
x = np.linspace(-10, 10, 100)

def f(x):
    return exp( -exp(x) )

y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the function y =  exp(-exp(x))')
plt.grid(True)
plt.show()
```
